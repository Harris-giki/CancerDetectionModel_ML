```python
  import numpy as np
  import pandas as pd
  import matplotlib.pyplot as plt
  import sklearn.datasets
  from sklearn.model_selection import train_test_split
```

#DATA COLLECTION AND PROCESSING

```python
#loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
print(breast_cancer_dataset) #the data is stored here as dictionaries
```

```python
#loading the data to DataFrame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
data_frame.head()
```

The number of coloumn were specified as the features in the dataset, therefore there are 30 features in the imported dataset.

```python
# adding the 'target column' to the data frame
data_frame['label']=breast_cancer_dataset.target
#printing last 5 rows of the DataFrame
data_frame.tail()
```

Previously there was no column specificing the dataset results to be 'Malignant' or 'Benign'. The above mentioned code would specify the results/target (as 0/1) in a new column named 'Label'

```python
from re import X
#initialzing separate variables for the freatures and targets
X = data_frame.drop(columns=['label'], axis=1)
Y = data_frame['label']
print(X)
print(Y)
```

```python
#Spliting the dataset into training and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Standardizing the dataset enhances the learning model
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)

print(X_train_std)
print(X_test_std)
```

#BUILDING THE NEURAL NETWORK

```python
#importing relevant libraries
import tensorflow as tf
#when we are creating neural networks there are some weights and parameters that are initialized
#seeding the value would generate the random values accordingly without changing the outputs everytime the code restarts
tf.random.set_seed(3)

from tensorflow import keras
```

```python
# setting up the layers of Neural Network
# sequential is a keras function that is used to stack the layers of the neural networks
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)), # as the input layer accepts a 1D array, we convert the matrix into 1D array by this function
    keras.layers.Dense(20, activation='relu'),# the hidden layer
    keras.layers.Dense(2, activation='sigmoid')# the output layer
])
```

```python
# compiling the neural network
model.compile(optimizer='adam',
loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
#training the Neural Network
history = model.fit(X_train, Y_train, validation_split=0.1, epochs=15) #epochs is the number of time our model goes through the dataset
```

```python
# visualizing accuracy and loss

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='lower right')
```

Now testing accuracy of the model on the 'Test Data'


```python
loss, accuracy = model.evaluate(X_test, Y_test)
print("Test Loss:", loss)
print("Test Accuracy:", accuracy)
```
